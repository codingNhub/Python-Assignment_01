```python
import matplotlib as plt
```

```python
from matplotlib import pyplot
```

```python
df = pd.read_csv('jsrt_metadata.csv')
```

```python

print(df['diagnosis'].value_counts())
```

```python
print("Total number of patients of each disease:",df['diagnosis'].value_counts().sum())
```

```python
df['diagnosis']=df['diagnosis'].astype('str')
```

```python
plt.pyplot.hist(df['diagnosis'].value_counts(),color='purple')
```

```python
plt.pyplot.title('Total number of patients of each disease:')
```

```python
plt.pyplot.ylabel('Frequency')
```

```python
plt.pyplot.xlabel('Category')
```

```python
#2. Find the total number of patients of each disease (diagnosis) with respect to the disease state(malignant or benign) and plot in a pie chart.
plt.pyplot.pie(df[['diagnosis','state']].value_counts(),shadow=True
,colors=['purple','pink','orange'],autopct='%1f%%',startangle=0,radius=2)
```

```python
#3.Find and plot via pia chart the total number of patients of each disease (diagnosis) with respect to the gender(Female,Male)

plt.pyplot.pie(df[['diagnosis','gender']].value_counts(),shadow=True
,colors=['purple','pink','orange'],autopct='%1f%%',startangle=0,radius=2)
```

```python
plt.pyplot.pie(df['state'].value_counts(),shadow=True
,colors=['purple','pink','orange'],autopct='%1f%%',startangle=0,radius=2)
```
